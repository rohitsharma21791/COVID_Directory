```python
# -*- coding: utf-8 -*-
"""
Created on Sat Aug  7 23:33:36 2021

@author: Sharma
"""


from keras.layers import Input
from keras.layers import Input, Dense, Conv2D, Conv3D, DepthwiseConv2D, SeparableConv2D, Conv3DTranspose
from keras.layers import Flatten, MaxPool2D, AvgPool2D, GlobalAvgPool2D, UpSampling2D, BatchNormalization
from keras.layers import Concatenate, Add, Dropout, ReLU, Lambda, Activation, LeakyReLU, PReLU, add




from time import time
import numpy as np
import tensorflow as tf


from keras.models import Model
from tensorflow.keras import initializers
kernel_initializer=initializers.RandomNormal(stddev=0.01)
def h_sigmoid(x):
    return tf.nn.relu6(x + 3) / 6

def h_swish(x):
    return x * h_sigmoid(x)

def mobile_netv3(input_shape,n_classes):
    
    def se_block(x1,input_channels,r=16):
        x=GlobalAvgPool2D()(x1)
        x=Dense(units=input_channels//r)(x)
        x=ReLU()(x)
        x=Dense(units=input_channels)(x)
        x=h_sigmoid(x)
        x=tf.expand_dims(input=x, axis=1)
        x=tf.expand_dims(input=x, axis=1)
        
        x=x1*x
        return x

    def se_block1(x1,input_channels,r):
      num_reduced_filters = max(1, int(input_channels//r))
      x=GlobalAvgPool2D()(x1)
      x=tf.expand_dims(input=x,axis=1)
      x=tf.expand_dims(input=x,axis=1)
      x=Conv2D(filters=num_reduced_filters,kernel_size=(1,1),strides=1,padding="same",kernel_initializer=kernel_initializer)(x)
      x=h_sigmoid(x)
      x=Conv2D(filters=input_channels,kernel_size=(1,1),strides=1,padding="same",kernel_initializer=kernel_initializer)(x)     
      x=h_sigmoid(x)
      x=x1*x
      return x

    

    
    def BottleNeck1(x1,in_size,exp_size,out_size,s,is_se_existing,NL,k,r):
        
        
        if NL=="HS":
            x=h_swish(x1)
        elif NL=="RE":
            x=tf.nn.relu6(x1)
        x=DepthwiseConv2D(kernel_size=(k, k),strides=s,padding="same",use_bias=False)(x)
        x=BatchNormalization()(x)
        if NL=="HS":
            x=h_swish(x)
        elif NL=="RE":
            x=tf.nn.relu6(x)
        if is_se_existing:
            x=se_block1(x,input_channels=exp_size,r=r)
        x=Conv2D(filters=out_size,kernel_size=(1, 1),strides=1,padding="same",use_bias=False)(x)
        x=BatchNormalization()(x)
        x=Activation(tf.keras.activations.linear)(x)
        if s==1 and in_size==out_size:
            x=add([x,x1])
        return x


    
    def BottleNeck(x1,in_size,exp_size,out_size,s,is_se_existing,NL,k,r):
        x=Conv2D(filters=exp_size,kernel_size=(1, 1),strides=1,padding="same",use_bias=False)(x1)
        x=BatchNormalization()(x)
        
        if NL=="HS":
            x=h_swish(x)
        elif NL=="RE":
            x=tf.nn.relu6(x)
        x=DepthwiseConv2D(kernel_size=(k, k),strides=s,padding="same",use_bias=False)(x)
        x=BatchNormalization()(x)
        if NL=="HS":
            x=h_swish(x)
        elif NL=="RE":
            x=tf.nn.relu6(x)
        if is_se_existing:
            x=se_block1(x,input_channels=exp_size,r=r)
        x=Conv2D(filters=out_size,kernel_size=(1, 1),strides=1,padding="same",use_bias=False)(x)
        x=BatchNormalization()(x)
        x=Activation(tf.keras.activations.linear)(x)
        if s==1 and in_size==out_size:
            x=add([x,x1])
        return x
    input=Input(input_shape)
    x=Conv2D(filters=16,kernel_size=(3,3),strides=2,padding="same",use_bias=False)(input)
    x=BatchNormalization()(x)
    x=BottleNeck1(x,in_size=16, exp_size=16, out_size=16, s=2, is_se_existing=True, NL="RE", k=3,r=2)
    x=BottleNeck(x,in_size=16, exp_size=72, out_size=24, s=2, is_se_existing=False, NL="RE", k=3,r=2)
    x=BottleNeck(x,in_size=24, exp_size=88, out_size=24, s=1, is_se_existing=False, NL="RE", k=3,r=2)
    x=BottleNeck(x,in_size=24, exp_size=96, out_size=40, s=2, is_se_existing=True, NL="HS", k=5,r=4)
    x=BottleNeck(x,in_size=40, exp_size=240, out_size=40, s=1, is_se_existing=True, NL="HS", k=5,r=4) #3.75
    x=BottleNeck(x,in_size=40, exp_size=240, out_size=40, s=1, is_se_existing=True, NL="HS", k=5,r=4)   #3.75
    x=BottleNeck(x,in_size=40, exp_size=120, out_size=48, s=1, is_se_existing=True, NL="HS", k=5,r=4)  #3.75
    x=BottleNeck(x,in_size=48, exp_size=144, out_size=48, s=1, is_se_existing=True, NL="HS", k=5,r=4)      #144/41
    x=BottleNeck(x,in_size=48, exp_size=288, out_size=96, s=2, is_se_existing=True, NL="HS", k=5,r=4)
    x=BottleNeck(x,in_size=96, exp_size=576, out_size=96, s=1, is_se_existing=True, NL="HS", k=5,r=4)
    x=BottleNeck(x,in_size=96, exp_size=576, out_size=96, s=1, is_se_existing=True, NL="HS", k=5,r=4)
    x=Conv2D(filters=576,kernel_size=(1,1),strides=1,padding="same",use_bias=False)(x)
    x=BatchNormalization()(x)
    x=AvgPool2D(pool_size=(7, 7),strides=1)(x)
    x=Conv2D(filters=1024,kernel_size=(1,1),strides=1,padding="same")(x)
    output=Conv2D(filters=n_classes,kernel_size=(1,1),strides=1,padding="same",activation=tf.keras.activations.sigmoid)(x)
    
    model = Model(input, output)
    return model

    
    
input_shape=224,224,1
n_classes=1

model = mobile_netv3(input_shape,n_classes)
model.summary()
```

```python
from tensorflow.keras.optimizers import SGD
epochs=100
learning_rate = 0.01
decay_rate = learning_rate / epochs
momentum = 0.8
sgd = SGD(learning_rate=learning_rate, momentum=momentum, decay=decay_rate, nesterov=True)




def exp_decay(epoch):
    lrate = learning_rate * np.exp(-decay_rate*epoch)
    return lrate

# learning schedule callback
lr_rate = LearningRateScheduler(exp_decay)
#Default values for SGD. lr=0.1, m=0, decay=0
#Nesterov momentum is a different version of the momentum method.
#Nesterov has stronger theoretical converge guarantees for convex functions.
```

```python
################# Ideal Theshold Calculation ############################

y_prob=model.predict(X_test).ravel()
y_test1=np.array(y_test).ravel()

import numpy as np

def roc_curve(y_test1, y_prob, thresholds):

    fpr = []
    tpr = []
    
    for threshold in thresholds:

        y_pred = np.where(y_prob >= threshold, 1, 0)

        fp = np.sum((y_pred == 1) & (y_test1== 0))
        tp = np.sum((y_pred == 1) & (y_test1 == 1))

        fn = np.sum((y_pred == 0) & (y_test1 == 1))
        tn = np.sum((y_pred == 0) & (y_test1 == 0))

        fpr.append(fp / (fp + tn))
        tpr.append(tp / (tp + fn))

    return [fpr, tpr]

y_prob1 = model.predict(X_test).ravel()
thresholds1= list(np.arange(-100, 100, 0.001))
[fpr1,tpr1]=roc_curve(y_test1,y_prob1, thresholds1)
thresholds1=np.array(thresholds1)
tpr1=np.array(tpr1)
fpr1=np.array(fpr1)
import pandas as pd
i = np.arange(len(tpr1)) 
roc1 = pd.DataFrame({'tf' : pd.Series(tpr1-(1-fpr1), index=i), 'thresholds1' : pd.Series(thresholds1, index=i)})
ideal_roc_thresh1 = roc1.iloc[(roc1.tf-0).abs().argsort()[:1]]  #Locate the point where the value is close to 0
print("Ideal threshold is: ", ideal_roc_thresh1['thresholds1'])
```
